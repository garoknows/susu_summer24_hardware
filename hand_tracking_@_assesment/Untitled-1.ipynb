```python
import os
import ctypes
import vlc
import time
import keyboard

# Set the path to the VLC installation directory
vlc_path = r'C:/Program Files/VideoLAN/VLC'

# Ensure the VLC path exists
if not os.path.exists(vlc_path):
    raise FileNotFoundError(f"VLC path not found: {vlc_path}")

# Add the VLC directory to the DLL search path
os.add_dll_directory(vlc_path)

# Set the full path to the libvlc.dll file
libvlc_dll = os.path.join(vlc_path, 'libvlc.dll')

# Load the libvlc.dll with error handling
try:
    ctypes.CDLL(libvlc_dll)
except OSError as e:
    raise OSError(f"Error loading libvlc.dll: {e}")

# Create an instance of VLC
instance = vlc.Instance()

# Create a media player object
player = instance.media_player_new()

# Define the webcam media
webcam_device = 'dshow://'  # For Windows

def start_recording():
    global player, recording
    record_start_time = time.strftime('%Y-%m-%d_%H-%M-%S')
    output_file = os.path.join(os.getcwd(), f"recording_{record_start_time}.mp4")
    
    # HEVC encoding options
    options = (
        ":sout=#duplicate{dst=display,dst=standard{access=file,mux=mp4,venc=x265{preset=medium,tune=zerolatency},dst=" +
        output_file +
        "}},sout-ffmpeg-strict=-2"
    )
    
    media = instance.media_new(webcam_device, options)
    player.set_media(media)
    player.play()
    recording = True
    print(f"Recording started: {output_file}")

def stop_recording():
    global player, recording
    player.stop()
    recording = False
    print("Recording stopped")
    # Reset media to display-only mode
    media = instance.media_new(webcam_device)
    player.set_media(media)
    player.play()

# Start playing the webcam stream
player.play()

recording = False

# Keep the stream open and manage recording with 'r' key
try:
    print("Press 'r' to start/stop recording, 'q' to quit")
    while True:
        if keyboard.is_pressed('q'):
            print("Quitting...")
            break

        if keyboard.is_pressed('r'):
            if not recording:
                start_recording()
            else:
                stop_recording()
            # Debounce the key press
            time.sleep(1)

        time.sleep(0.1)
except KeyboardInterrupt:
    pass

# Stop the media player
player.stop()
```
